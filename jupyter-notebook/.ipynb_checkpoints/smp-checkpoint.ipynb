```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import plotly as py
import plotly.graph_objs as go
from plotly.offline import plot

from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
init_notebook_mode(connected=True) 
```

```python
tesla = pd.read_csv('../jupyter-notebook/tesla.csv')
tesla.head()
```

```python
tesla.info()
```

```python
tesla['Date'] = pd.to_datetime(tesla['Date'],dayfirst=True)
```

```python
print(f'Dataframe contains stock prices between {tesla.Date.min()} {tesla.Date.max()}') 
print(f'Total days = {(tesla.Date.max() - tesla.Date.min()).days} days')
```

```python
tesla.describe()
```

```python
tesla[['Open','High','Low','Close','Adj Close']].plot(kind='box')
```

```python
layout = go.Layout(
    title='Stock Prices of Tesla',
    xaxis=dict(
        title='Date',
        titlefont=dict(
            family='Courier New, monospace',
            size=18,
            color='#7f7f7f'
        )
    ),
    yaxis=dict(
        title='Price',
        titlefont=dict(
            family='Courier New, monospace',
            size=18,
            color='#7f7f7f'
        )
    )
)

tesla_data = [{'x':tesla['Date'], 'y':tesla['Close']}]
plot = go.Figure(data=tesla_data, layout=layout)
```

```python
iplot(plot)
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(tesla['Volume'], tesla['Close'], alpha=0.5, color='purple')
plt.title('Volume vs. Closing Price')
plt.xlabel('Volume')
plt.ylabel('Closing Price')
plt.grid(True)
plt.show()
```

```python
import seaborn as sns
```

```python
tesla['50_MA'] = tesla['Close'].rolling(window=50).mean()
tesla['200_MA'] = tesla['Close'].rolling(window=200).mean()
plt.plot(tesla['Date'], tesla['Close'], label='Closing Price')
plt.plot(tesla['Date'], tesla['50_MA'], label='50-Day MA', linestyle='--')
plt.plot(tesla['Date'], tesla['200_MA'], label='200-Day MA', linestyle='--')
```

```python
plt.figure(figsize=(12, 6))
plt.plot(tesla['Date'], tesla['Volume'], label='Volume', color='Teal')
plt.xlabel('Date')
plt.ylabel('Trading Volume')
plt.title('Trading Volume Over Time')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error as mse
from sklearn.metrics import r2_score
```

```python
X = np.array(tesla.index).reshape(-1,1)
Y = tesla['Close']
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=101)
```

```python
scaler = StandardScaler().fit(X_train)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
lm.fit(X_train, Y_train)
```

```python
trace0 = go.Scatter(
    x = X_train.T[0],
    y = Y_train,
    mode = 'markers',
    name = 'Actual'
)
trace1 = go.Scatter(
    x = X_train.T[0],
    y = lm.predict(X_train).T,
    mode = 'lines',
    name = 'Predicted'
)
tesla_data = [trace0,trace1]
layout.xaxis.title.text = 'Day'
plot2 = go.Figure(data=tesla_data, layout=layout)
```

```python
iplot(plot2)
```

```python
scores = f'''
{'Metric'.ljust(10)}{'Train'.center(20)}{'Test'.center(20)}
{'r2_score'.ljust(10)}{r2_score(Y_train, lm.predict(X_train))}\t{r2_score(Y_test, lm.predict(X_test))}
{'MSE'.ljust(10)}{mse(Y_train, lm.predict(X_train))}\t{mse(Y_test, lm.predict(X_test))}
'''
print(scores)
```
