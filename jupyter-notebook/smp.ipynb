```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error as mse
from sklearn.metrics import r2_score
```

```python
X = np.array(tesla.index).reshape(-1,1)
Y = tesla['Close']
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=101)
```

```python
X
```

```python
scaler = StandardScaler().fit(X_train)
X_train=scaler.transform(X_train)
```

```python
X_train
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
lm.fit(X_train, Y_train)
```

```python
trace0 = go.Scatter(
    x = X_train.T[0],
    y = Y_train,
    mode = 'markers',
    name = 'Actual'
)
trace1 = go.Scatter(
    x = X_train.T[0],
    y = lm.predict(X_train).T,
    mode = 'lines',
    name = 'Predicted'
)
tesla_data = [trace0,trace1]
layout.xaxis.title.text = 'Day'
plot2 = go.Figure(data=tesla_data, layout=layout)
```

```python
scores = f'''
{'Metric'.ljust(10)}{'Train'.center(20)}{'Test'.center(20)}
{'r2_score'.ljust(10)}{r2_score(Y_train, lm.predict(X_train))}\t{r2_score(Y_test, lm.predict(X_test))}
{'MSE'.ljust(10)}{mse(Y_train, lm.predict(X_train))}\t{mse(Y_test, lm.predict(X_test))}
'''

print(scores)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor
from sklearn.feature_selection import SelectFromModel
```

```python
features = ['Open', 'High', 'Low', 'Volume']
X = tesla[features].values
y = tesla['Close'].values

rf_model = RandomForestRegressor(random_state=42)
```

```python
param_grid = {
    'n_estimators': [50, 100, 150],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

grid_search = GridSearchCV(estimator=rf_model, param_grid=param_grid, cv=5, n_jobs=-1, verbose=2)
grid_search.fit(X_train, Y_train)

best_rf_model = grid_search.best_estimator_

sfm = SelectFromModel(best_rf_model, prefit=True)
selected_features = [features[i] for i in sfm.get_support(indices=True)]

X_train_selected = sfm.transform(X_train)
X_test_selected = sfm.transform(X_test)

best_rf_model.fit(X_train_selected, Y_train)

y_pred = best_rf_model.predict(X_test_selected)
```

```python
ms = mse(Y_test, y_pred)
r2 = r2_score(Y_test, y_pred)

print("Best Random Forest Model:")
print(f"Mean Squared Error (MSE): {ms}")
print(f"R-squared (R2) Score: {r2}")
print(f"Selected Features: {selected_features}")
```
